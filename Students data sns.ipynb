```python
import pandas as pd 
import seaborn as sns 
```

# Seaborn data visualization 

```python
Df = pd.read_csv("student-dataset.csv")
Df.head()
```

# sns box plots

```python
sns.boxplot(y = "math.grade", data = Df) 
```

```python
sns.boxplot(x = "english.grade",y = "math.grade", data = Df) 
```

# sns violin plot

```python
sns.violinplot(x = "math.grade",y = "english.grade", data = Df) 
```

# reg plot

```python
sns.regplot(x = "math.grade",y = "english.grade", data = Df) 
```
